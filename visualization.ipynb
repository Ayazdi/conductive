```python
import plotly.figure_factory as ff
from scipy.spatial import Delaunay


points2D = np.vstack([site['SITE_X'], site['SITE_Y']]).T
tri = Delaunay(points2D)
simplices = tri.simplices
fig = ff.create_trisurf(site['SITE_X'], site['SITE_Y'], site['SITE_Z'],
                         simplices=simplices,
                         title="wafare", aspectratio=dict(x=1, y=1, z=0.5))
fig.show()
```

```python
from scipy.interpolate import griddata, LinearNDInterpolator, interp2d
import plotly.figure_factory as ff
from PIL import Image

x = site['SITE_X']
y = site['SITE_Y']
z = site['SITE_Z']
# points2D = np.vstack([site['SITE_X'], site['SITE_Y']]).T
cartcoord = list(zip(x, y))
X = np.linspace(min(x), max(x))
Y = np.linspace(min(y), max(y))
X, Y = np.meshgrid(X, Y)

# Approach 1
interp = LinearNDInterpolator(cartcoord, z, fill_value=0)
Z0 = interp(X, Y)

fig = ff.create_trisurf(X, Y, Z0,
                         simplices=simplices,
                         title="wafare", aspectratio=dict(x=1, y=1, z=0.5))
fig.show()

plt.figure()
plt.pcolormesh(X, Y, Z0)
plt.colorbar() # Color Bar
plt.show()

```

```python
# Approach 2
func = interp2d(x, y, z)
Z = func(X[0, :], Y[:, 0])
plt.figure()
plt.pcolormesh(X, Y, Z)
plt.colorbar() # Color Bar
plt.show()
```
