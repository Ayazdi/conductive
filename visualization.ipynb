## Reconstruction of wafer profile using interpolation.

```python
import plotly.figure_factory as ff
from scipy.spatial import Delaunay

def visualize(thickness, title):
    site['SITE_Z'] = thickness
    points2D = np.vstack([site['SITE_X'], site['SITE_Y']]).T
    tri = Delaunay(points2D)
    simplices = tri.simplices

    fig = ff.create_trisurf(site['SITE_X'], site['SITE_Y'], site['SITE_Z'],
                             simplices=simplices,
                             title=title, aspectratio=dict(x=1, y=1, z=0.5))
    fig.show()
```

```python
tool_1 = df[all_sites].iloc[0].values
tool_2 = df[all_sites].iloc[1].values
tool_3 = df[all_sites].iloc[2].values
tool_4 = df[all_sites].iloc[3].values

visualize(tool_1, 'flow: 0.916 space:0.344 dep_time: 69.289 tool:1' )
visualize(tool_2, 'flow: 0.916 space:0.344 dep_time: 69.289 tool:2' )
visualize(tool_3, 'flow: 0.916 space:0.344 dep_time: 69.289 tool:3' )
visualize(tool_4, 'flow: 0.916 space:0.344 dep_time: 69.289 tool:4' )
```

## Creating augmented data:

In order to include extrapolation of 1 standard deviation outside of the ranges of parameters and increase our data set augmented data was generated. From each sample, 30 samples with the same thickness profile were created and the values of the parameter (except tool) were generated from a normal distribution with a mean of the parameter value and total std of the parameter.

```python
# standard deviation
flow_std = np.std(df['FLOWFACTOR'])
space_std = np.std(df['SPACING'])
dep_std = np.std(df['DEP TIME'])

for i in range(0, 500):
    # creating normal distribution of each parameter with the total std of that the parameter
    aug_flow = np.random.normal(loc=df.loc[i,'FLOWFACTOR'], scale=flow_std, size=30) 
    aug_spacing = np.random.normal(loc=df.loc[i,'SPACING'], scale=space_std, size=30)
    aug_dep = np.random.normal(loc=df.loc[i,'DEP TIME'], scale=dep_std, size=30)
    
    # not changing the tool
    aug_tool = np.zeros(30)
    aug_tool = df.loc[i,'TOOL']

    aug_dict = {'FLOWFACTOR':aug_flow, 'SPACING':aug_spacing, 'DEP TIME':aug_dep, 'TOOL':aug_tool}


    aug_df = pd.DataFrame(aug_dict)
    # adding the augmented data to the main df
    site_col = pd.DataFrame(columns=all_sites)
    aug_df = pd.concat([aug_df, site_col], axis=1)
    aug_df.loc[:, all_sites] = df[all_sites].iloc[i].values
    df = pd.concat([df, aug_df], axis=0)
    df.reset_index(drop=True, inplace=True)
df
```

```python
# saving the new dataset
df.to_csv('aug_df.csv')
```
